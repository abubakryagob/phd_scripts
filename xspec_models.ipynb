```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Generate some data for the errorbar plots
x = np.array([1, 2, 3, 4])
y1 = np.array([4, 3, 5, 2])
y2 = np.array([3, 5, 2, 4])
y3 = np.array([5, 2, 4, 3])
error1 = np.array([0.5, 0.3, 0.7, 0.2])
error2 = np.array([0.3, 0.7, 0.2, 0.5])
error3 = np.array([0.7, 0.2, 0.5, 0.3])

# Create a figure with a grid of three rows and one column, with the bottom two rows sharing the x-axis
fig = plt.figure(figsize=(12, 6))
gs = gridspec.GridSpec(3, 1, figure=fig, height_ratios=[1, 1, 1], hspace=0.1)
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1], sharex=ax1)
ax3 = fig.add_subplot(gs[2], sharex=ax1)

# Plot the errorbar data in each panel
ax1.errorbar(x, y1, yerr=error1, fmt='o')
ax2.errorbar(x, y2, yerr=error2, fmt='o')
ax3.errorbar(x, y3, yerr=error3, fmt='o')

# Remove x-tick labels for the top two panels
ax1.tick_params(labelbottom=False)
ax2.tick_params(labelbottom=False)

# Add titles and axis labels
ax1.set_title('Panel 1')
ax2.set_title('Panel 2')
ax3.set_title('Panel 3')
fig.suptitle('Three Panels with Errorbar Plots')
fig.text(0.5, 0.04, 'X axis', ha='center')
fig.text(0.04, 0.5, 'Y axis', va='center', rotation='vertical')

# Display the figure
plt.show()
```
